Заполним столбцы 'movie_info', 'critic_consensus' значениями "unknown" тк невозможно подобрать значения на основе остальных столбцов

```python
columns_to_replace =['movie_info', 'critic_consensus'
       ]

for row in columns_to_replace:
    df[row] = df[row].fillna('unknown')
```

Проанализировав данные можно понять, что почти все строки, которые не идеют режиссера, так же не имеют всех остальных значений. 
Поэтому их можно удалить.
Заполним "unknown" и удалим

```python
columns_to_replace =['directors'
       ]

for row in columns_to_replace:
    df[row] = df[row].fillna('unknown')
    
index_names = df[ df['directors'] == 'unknown' ].index
df.drop(index_names, inplace = True)
df.isna().sum()
```

Остальные данные заполним основываясь на студии создания фильма

```python
x = lambda x: x.mode().iat[0] if x.notna().any() else np.nan
columns_w_nan = {'authors', 'streaming_release_date', 
       'audience_status'}
for column_nan in columns_w_nan:
    col = df.groupby('film_studio_producer')[column_nan].transform(x)
    df [column_nan] = df [column_nan].fillna(col)
```

```python
df.isna().sum()
```

На самом деле эти пустые ячейки не заполнятся ни по одному из критериев, поэтому их тоже удалим. 3 удаленных строки не сильно повлияют на датафрейм

```python
columns_to_replace =['audience_status', 'authors', 'streaming_release_date'
       ]

for row in columns_to_replace:
    df[row] = df[row].fillna('unknown')


for row in columns_to_replace:
    index_names = df[ df[row] == 'unknown' ].index
    df.drop(index_names, inplace = True)
df.isna().sum()
```

```python
## Изучение и обработка выбросов
```

```python
outlier = df[['avg_rate','film_age','tomatometer_rating','audience_rating', 'tomatometer_status']]
outlier
```

```python
# Определить нижние пределы для выбросов
Q1 = outlier.quantile(0.25)
Q1
```

```python
# Определить верхние пределы выбросов
Q3 = outlier.quantile(0.75)
Q3
```

```python
# IQR-оценки
IQR = Q3-Q1
# Отфильтрованные данные, из которых удалены выбросы
data_filtered = outlier[~((outlier < (Q1 - 1.5 * IQR)) |(outlier > (Q3 + 1.5 * IQR))).any(axis=1)]
data_filtered
```

```python
# Копирование индексов
index_list = list(data_filtered.index.values)
```

```python
# Хранить данные без выбросов в отдельном DataFrame.
df_filtered = df[df.index.isin(index_list)]
df_filtered.head()
```

```python
# создать гистограмму с отфильтрованным выбросом
filtered_outlier = df_filtered[['avg_rate','film_age','tomatometer_rating','audience_rating', 'tomatometer_status']]

for col in filtered_outlier:
    df_filtered.hist(col, bins=100)
```

## Рейтинг

Рассмотрим факторы которые влияют на рейтинг фильмов от двух самых больших студий.

- Возраст
- Время
- Награда
- Статусы на различных сайтах
- Возратной рейтинг

Для категориальных переменных (Награда, Статус) использовались прямоугольные диаграммы, а для остальных — диаграммы рассеяния.
Матрица корреляции и графики корреляции будут использоваться для оценки силы и направления линейных связей между парами переменных.

```python
Paramount = data_filtered.query('production_company == "Paramount Pictures"')
Warner = data_filtered.query('production_company == "Warner Bros. Pictures"')
```

```python
sns.lmplot(x="film_age", y="avg_rate", data=Paramount)
plt.title('Paramount Pictures')
sns.lmplot(x="film_age", y="avg_rate", data=Warner)
plt.title('Warner Bros.')
```

```python
age_Paramount = Paramount[['film_age','avg_rate']]
corr_Paramount = age_Paramount.corr()
age_Warner = Warner[['film_age','avg_rate']]
corr_Warner = age_Warner.corr()
```

```python
round(corr_Paramount,3)
```

```python
round(corr_Warner,3)
```

**Заключение** :
<br>
Видно, что у студии Paramount рейтинг фильмов увеличивается с возрастом, а у Ворнер наоборот. Это значит, что первая студия раньше выпускала фильмы лучше, а вторая учится на ошибках и выпускает фильмы лучше со временем

```python
sns.lmplot(x="movie_time", y="avg_rate", data=Paramount)
plt.title('Paramount')
sns.lmplot(x="movie_time", y="avg_rate", data=Warner)
plt.title('Warner Bros.')
```

```python
age_Paramount = Paramount[['movie_time','avg_rate']]
corr_Paramount = age_Paramount.corr()
age_Warner = Warner[['movie_time','avg_rate']]
corr_Warner = age_Warner.corr()
```

```python
round(corr_Paramount,3)
```

```python
round(corr_Warner,3)
```

**Заключение** :
<br>
Здесь же видно, что у Paramount пользователям без разницы сколько будет идти фильм.
А вот у Ворнер чем длиннее фильм, тем он скучнее и больше не нравится зрителям и критикам.
